```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.worldatlas.com/countries"
```

```python
response = requests.get(url)
```

```python
print(response)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Extracting Country Codes.
code = soup.find_all('td', {'id' : 'item_iso3'})
```

```python
# Creating a list of all country codes.
country_code = []

for code in code:
    country_code.append(code.text)
```

```python
country_code
```

```python
# Extracting the image URLs.
image = soup.find_all('img')
```

```python
# Creating a list of all country flag image URLs.
country_flag = []

for flag in image[:-1]:
    country_flag.append("https://www.worldatlas.com" + flag['src'])
```

```python
country_flag
```

```python
# Creating a dictionary so that we can merge all the lists and convert them into a DataFrame.
flag_dict = {
    'Country Code' : country_code,
    'Country Flag' : country_flag
    
}
```

```python
#Creating a DataFrame.
flag_df = pd.DataFrame(flag_dict)
```

```python
flag_df
```

```python
#Loading Country Data file 
Country_data = pd.read_csv("Country Data.csv")
```

```python
Country_data
```

```python
merged_df = pd.merge(Country_data, flag_df[["Country Code","Country Flag"]], 
                                            left_on = "Code",
                                            right_on = "Country Code",
                                            how ='left')
```

```python
merged_df
```

```python

merged_df = merged_df.drop('Country Code', axis=1)

# rename the 'Code' column to 'Country Code'
merged_df = merged_df.rename(columns={'Code': 'Country Code'})
```

```python
merged_df
```

```python
merged_df.to_csv("Merged Country Data.csv",index=False)
```
